### joins

```python
course1 = {'stdid':[1,2,3,4],'course1':[78,80,90,85]}
course2 = {'stdid':[2,3,4,5],'course1':[88,90,88,93]}
dfc1 = pd.DataFrame(course1)
dfc2 = pd.DataFrame(course2)
dfc2
```

### inner join

```python
gradesdf = dfc1.merge(dfc2,on=['stdid'])
gradesdf
```

### left,right and full outer join

```python
gradesdf2 = dfc1.merge(dfc2,on='stdid',how='left')
gradesdf2
```

```python
gradesdf3 = dfc1.merge(dfc2,on='stdid',how='right')
gradesdf3
```

```python
gradesdf4 = dfc1.merge(dfc2,on='stdid',how='outer')
gradesdf4
```
