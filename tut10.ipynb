```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import random
import pandas as pd
```

```python
np.random.seed(2)
pageSpeed = np.random.normal(3.0,1.0,100)
purchaseAmount = np.random.normal(50.0,30.0,100)/pageSpeed
plt.scatter(pageSpeed,purchaseAmount)
```

```python
%matplotlib notebook
trainX = pageSpeed[:80]
testX = pageSpeed[80:]

trainY = purchaseAmount[:80]
testY = purchaseAmount[80:]

plt.scatter(trainX,trainY)
plt.show
```

```python
%matplotlib notebook
plt.scatter(testX,testY)
plt.show
```

```python
%matplotlib notebook
x = np.array(trainX)
y = np.array(trainY)

p4 = np.poly1d(np.polyfit(x,y,14))
xp = np.linspace(0,7,100)
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0,200])
plt.scatter(x,y)
plt.plot(xp,p4(xp),c='r')
plt.show()
```

```python
%matplotlib notebook
x = np.array(testX)
y = np.array(testY)

axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0,200])
plt.scatter(testX,testY)
plt.plot(xp,p4(xp),c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(testY,p4(testX))
print(r2)
```
